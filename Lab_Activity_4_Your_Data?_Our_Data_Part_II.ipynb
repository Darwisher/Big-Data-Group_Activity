```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import col, desc

spark = SparkSession.builder \
    .appName("NBA Player Stats Analysis") \
    .getOrCreate()
```

```python
# Define the structure and data types of the dataset
schema = StructType([
    StructField("Rk", IntegerType(), True),
    StructField("Player", StringType(), True),
    StructField("Pos", StringType(), True),
    StructField("Age", IntegerType(), True),
    StructField("Tm", StringType(), True),
    StructField("G", IntegerType(), True),
    StructField("GS", IntegerType(), True),
    StructField("MP", FloatType(), True),
    StructField("FG", FloatType(), True),
    StructField("FGA", FloatType(), True),
    StructField("FG_pct", FloatType(), True),
    StructField("3P", FloatType(), True),
    StructField("3PA", FloatType(), True),
    StructField("3P_pct", FloatType(), True),
    StructField("2P", FloatType(), True),
    StructField("2PA", FloatType(), True),
    StructField("2P_pct", FloatType(), True),
    StructField("eFG_pct", FloatType(), True),
    StructField("FT", FloatType(), True),
    StructField("FTA", FloatType(), True),
    StructField("FT_pct", FloatType(), True),
    StructField("ORB", FloatType(), True),
    StructField("DRB", FloatType(), True),
    StructField("TRB", FloatType(), True),
    StructField("AST", FloatType(), True),
    StructField("STL", FloatType(), True),
    StructField("BLK", FloatType(), True),
    StructField("TOV", FloatType(), True),
    StructField("PF", FloatType(), True),
    StructField("PTS", FloatType(), True)
])
```

```python
# Load the CSV dataset into a Spark DataFrame and shows first 5 rows of the dataset
df = spark.read.csv(
    "2021-2022 NBA Player Stats - Regular.csv",
    header=True,
    sep=";",
    schema=schema
)

df.show(5)
```

```python
# Display the DataFrame schema
df.printSchema()
```

```python
# Count the number of rows in the dataset
df.count()
```

```python
# Clean the dataset by removing duplicate and null records
df = df.dropDuplicates().dropna()
```

```python
# Get the top 5 players with the highest points per game
top_scorers = df.orderBy(desc("PTS")).select("Player","Tm","PTS").limit(5)

top_scorers.show()
```

```python
# Get the top 5 players with the highest assists
top_assists = df.orderBy(desc("AST")).select("Player","Tm","AST").limit(5)

top_assists.show()
```

```python
# Get the top 5 players with the highest rebounds
top_rebounds = df.orderBy(desc("TRB")).select("Player","Tm","TRB").limit(5)

top_rebounds.show()
```

```python
# Save the cleaned DataFrame into a Parquet file
df.write.mode("overwrite").parquet("nba_player_stats_clean.parquet")
```

```python
# Save the Insights into a Parquet file
top_scorers.write.mode("overwrite").parquet("top_scorers.parquet")
top_assists.write.mode("overwrite").parquet("top_assists.parquet")
top_rebounds.write.mode("overwrite").parquet("top_rebounds.parquet")
```
